Şehirlere göre PRICE ortalamasını inceleyelim.

```python
df.groupby(by=["SaleCityName","ConceptName"]).agg({"Price":"mean"})
```

 SaleCheckInDayDiff değişkenini kategorik değişkene çeviriyoruz

```python
bins=[-1,7,30,90,df["SaleCheckInDayDiff"].max()]
labels=["Last Minuters","Patential Planners","Planners","Early Bookers"]
df["EB_Scores"]=pd.cut(df["SaleCheckInDayDiff"],bins,labels=labels)
df.head(50).to_excel("eb_scorew.xlsx",index=False)
df.head(100)
```

Belli başlı kırılımlara göre ücret ortlamaları incelemesi yapıyoruz

```python
df.groupby(by=["SaleCityName","ConceptName","EB_Scores"]).agg({"Price":["mean","count"]})
```

```python

df.groupby(by=["SaleCityName","ConceptName","Seasons"]).agg({"Price":["mean","count"]})
```

```python
df.groupby(by=["SaleCityName","ConceptName","CInDay"]).agg({"Price":["mean","count"]})
```

City-Concept-Season kırılımının çıktısının PRICE'e göre sıralama işlemi

```python
agg_df=df.groupby(["SaleCityName","ConceptName","Seasons"]).agg({"Price":"mean"})
agg_df.head()
```

Indexte yer alan isimleri değişken ismine çeviriyoruz

```python
agg_df.reset_index(inplace=True)
agg_df.head(20)
```

Yeni seviye tabanlı müşterileri tanımlama işlemi

```python
agg_df['sales_level_based']=agg_df[["SaleCityName","ConceptName","Seasons"]].agg(lambda x:'_'.join(x).upper(),axis=1)
agg_df.head()
```

Yeni müşteriler için segmentlere ayırma işlemi

```python
agg_df["SEGMENT"]=pd.qcut(agg_df["Price"],4,labels=["D","C","B","A"])
agg_df.head(30)
```

```python
agg_df.groupby("SEGMENT").agg({"Price":["mean","max","sum"]})
```

Yeni gelen müşterileri sınıflandırıp ne kadar gelir getirebileceklerini tahmin ediyoruz.

```python
agg_df.sort_values(by="Price") #oluşan son df için price değişkenine göre sıralıyoruz
```
